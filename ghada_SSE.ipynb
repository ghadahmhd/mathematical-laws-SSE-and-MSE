```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
np.random.seed(42)

# Generate 100 random 2D points
X = np.random.rand(100, 2)

# Visualize the data
plt.scatter(X[:, 0], X[:, 1])
plt.title("Original Data")
plt.show()
```

```python
k = 3

kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
def compute_sse(X, labels, centroids):
    assigned_centroids = centroids[labels]
    squared_distances = np.sum((X - assigned_centroids) ** 2, axis=1)
    return np.sum(squared_distances)

sse = compute_sse(X, labels, centroids)
print("SSE =", sse)
```

```python
print("SSE from sklearn =", kmeans.inertia_)
```

```python
sse_values = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    sse_values.append(kmeans.inertia_)

plt.plot(range(1, 11), sse_values, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("SSE")
plt.title("Elbow Method")
plt.show()
```
